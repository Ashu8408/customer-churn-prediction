## 1. mean/percentage for features by churn status


```python
df = pd.read_csv('/content/drive/My Drive/raw_data/Telco-Customer-Churn.csv')

# Convert TotalCharges to numeric
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

cat_features = df.select_dtypes(include="object").columns.drop(["customerID", "Churn"])
num_features = df.select_dtypes(include=["int64", "float64"]).columns

# ---------------------------
# 1. Percentages for categorical features
# ---------------------------
cat_summary = {}
for col in cat_features:
    churn_rate = (
        df.groupby(col)["Churn"]
        .value_counts(normalize=True)
        .mul(100)
        .rename("percentage")
        .reset_index()
    )
    cat_summary[col] = churn_rate

# ---------------------------
# 2. Means for numerical features
# ---------------------------
num_summary = df.groupby("Churn")[num_features].mean().reset_index()

# ---------------------------
# Results
# ---------------------------
print("=== Numerical Feature Means by Churn ===")
print(num_summary)

print("\n=== Categorical Feature Percentages by Churn ===")
for col, table in cat_summary.items():
    print(f"\n{col}:")
    print(table)
```

## 1. mean/percentage for features by churn status
with graph


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load data
df = pd.read_csv('/content/drive/My Drive/raw_data/Telco-Customer-Churn.csv')

# Convert TotalCharges to numeric
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")

# Separate categorical and numerical features
cat_features = df.select_dtypes(include="object").columns.drop(["customerID", "Churn"])
num_features = df.select_dtypes(include=["int64", "float64"]).columns

# ---------------------------
# 1. Churn percentages by categorical features
# ---------------------------
for col in cat_features:
    churn_rate = (
        df.groupby(col)["Churn"]
        .value_counts(normalize=True)
        .mul(100)
        .rename("percentage")
        .reset_index()
    )
    plt.figure(figsize=(6, 4))
    ax = sns.barplot(
        data=churn_rate[churn_rate["Churn"] == "Yes"],
        x=col, y="percentage", palette="viridis", errorbar=None
    )

    # Annotate % values
    for p in ax.patches:
        ax.annotate(
            f"{p.get_height():.1f}%",                # format percentage
            (p.get_x() + p.get_width() / 2., p.get_height()), # position
            ha="center", va="bottom", fontsize=9, color="black"
        )

    plt.title(f"Churn Percentage by {col}")
    plt.xticks(rotation=45)
    plt.ylabel("Churn %")
    plt.tight_layout()
    plt.show()

# ---------------------------
# 2. Mean values of numerical features by churn
# ---------------------------
for col in num_features:
    plt.figure(figsize=(6, 4))
    ax = sns.barplot(
        data=df, x="Churn", y=col,
        hue="Churn", palette="coolwarm", estimator=np.mean, errorbar=None, legend=False
    )

    # Annotate mean values
    for p in ax.patches:
        ax.annotate(
            f"{p.get_height():.1f}",                # format numeric value
            (p.get_x() + p.get_width() / 2., p.get_height()),
            ha="center", va="bottom", fontsize=9, color="black"
        )

    plt.title(f"Mean {col} by Churn")
    plt.ylabel(f"Mean {col}")
    plt.tight_layout()
    plt.show()
```

# 2. bar graph, boxplot/histograms, correlation heatmap
without captions

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Separate categorical and numerical features
cat_features = df.select_dtypes(include="object").columns.drop(["customerID", "Churn"])
num_features = df.select_dtypes(include=["int64", "float64"]).columns

# ---------------------------
# 1. Bar plots for categorical features vs Churn
# ---------------------------
for col in cat_features:
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(data=df, x=col, hue="Churn", palette="Set2")
    plt.title(f"Churn Distribution by {col}")
    plt.xticks(rotation=45)

    # Annotate percentages on bars
    total = len(df)
    for p in ax.patches:
        count = int(p.get_height())
        percentage = 100 * count / total
        ax.annotate(
            f"{percentage:.1f}%",
            (p.get_x() + p.get_width() / 2., p.get_height()),
            ha="center", va="bottom", fontsize=8, color="black"
        )

    plt.tight_layout()
    plt.show()

# ---------------------------
# 2. Boxplots & Histograms for numerical features
# ---------------------------
for col in num_features:
    # Boxplot
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x="Churn", y=col, palette="coolwarm")
    plt.title(f"Boxplot of {col} by Churn")
    plt.tight_layout()
    plt.show()

    # Histogram
    plt.figure(figsize=(6, 4))
    ax = sns.histplot(
        data=df, x=col, hue="Churn", kde=True, element="step", palette="coolwarm"
    )
    plt.title(f"Distribution of {col} by Churn")

    # Annotate counts/percentages on histogram bins
    total = len(df)
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            percentage = 100 * height / total
            ax.annotate(
                f"{percentage:.1f}%",
                (p.get_x() + p.get_width() / 2., height),
                ha="center", va="bottom", fontsize=7, color="black", rotation=90
            )

    plt.tight_layout()
    plt.show()

# ---------------------------
# 3. Correlation Heatmap
# ---------------------------
plt.figure(figsize=(8, 6))
corr = df[num_features].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

# 3. bar graph, boxplot/histograms, correlation heatmap
with captions

```python
# ---------------------------
# 1. Bar plots for categorical features vs Churn
# ---------------------------
for col in cat_features:
    plt.figure(figsize=(6, 4))
    ax = sns.countplot(data=df, x=col, hue="Churn", palette="Set2")
    plt.title(f"Churn Distribution by {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Annotate counts + %
    total = len(df)
    for p in ax.patches:
        count = int(p.get_height())
        percentage = 100 * count / total
        ax.annotate(
            f"{count}\n({percentage:.1f}%)",
            (p.get_x() + p.get_width() / 2., p.get_height()),
            ha="center", va="bottom", fontsize=8, color="black"
        )

    # Add caption
    plt.figtext(
        0.5, -0.15,
        f"Caption: This shows how churn varies across different {col} categories.",
        wrap=True, horizontalalignment="center", fontsize=9
    )
    plt.show()

# ---------------------------
# 2. Boxplots & Histograms for numerical features
# ---------------------------
for col in num_features:
    # Boxplot
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x="Churn", y=col, palette="coolwarm")
    plt.title(f"Boxplot of {col} by Churn")
    plt.tight_layout()
    plt.figtext(
        0.5, -0.15,
        f"Caption: Compares the spread and median of {col} between churned vs non-churned customers.",
        wrap=True, horizontalalignment="center", fontsize=9
    )
    plt.show()

    # Histogram
    plt.figure(figsize=(6, 4))
    ax = sns.histplot(
        data=df, x=col, hue="Churn", kde=True, element="step", palette="coolwarm"
    )
    plt.title(f"Distribution of {col} by Churn")
    plt.tight_layout()

    # Annotate % on bins
    total = len(df)
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            percentage = 100 * height / total
            ax.annotate(
                f"{percentage:.1f}%",
                (p.get_x() + p.get_width() / 2., height),
                ha="center", va="bottom", fontsize=7, color="black", rotation=90
            )

    plt.figtext(
        0.5, -0.15,
        f"Caption: Shows how {col} values are distributed for churned vs retained customers.",
        wrap=True, horizontalalignment="center", fontsize=9
    )
    plt.show()

# ---------------------------
# 3. Correlation Heatmap
# ---------------------------
plt.figure(figsize=(8, 6))
corr = df[num_features].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.tight_layout()
plt.figtext(
    0.5, -0.05,
    "Caption: Highlights correlations between numerical features. High correlation may suggest redundancy.",
    wrap=True, horizontalalignment="center", fontsize=9
)
plt.show()
```
